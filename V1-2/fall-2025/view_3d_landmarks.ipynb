```python
#%pip install pyqt5
import matplotlib
matplotlib.use('Qt5Agg') 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mediapipe_extraction import main as main

def plot_3d_landmarks(unified_landmarks):
    """
    Create an interactive 3D plot of landmarks that you can rotate
    """
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    print(unified_landmarks)
    
    # Plot pose landmarks
    if unified_landmarks['pose']:
        pose_x = [p['x'] for p in unified_landmarks['pose']]
        pose_y = [p['y'] for p in unified_landmarks['pose']]
        pose_z = [p['z'] for p in unified_landmarks['pose']]
        ax.scatter(pose_x, pose_y, zs=pose_z, c='red', marker='o', s=50, label='Pose')
        ax.plot([pose_x[11], pose_x[12]], [pose_y[11], pose_y[12]], [pose_z[11], pose_z[12]], 'r-', alpha=0.6)
        ax.plot([pose_x[11], pose_x[13]], [pose_y[11], pose_y[13]], [pose_z[11], pose_z[13]], 'r-', alpha=0.6)
        ax.plot([pose_x[14], pose_x[12]], [pose_y[14], pose_y[12]], [pose_z[14], pose_z[12]], 'r-', alpha=0.6)
        ax.plot([pose_x[14], pose_x[16]], [pose_y[14], pose_y[16]], [pose_z[14], pose_z[16]], 'r-', alpha=0.6)
        ax.plot([pose_x[15], pose_x[13]], [pose_y[15], pose_y[13]], [pose_z[15], pose_z[13]], 'r-', alpha=0.6)
        ax.plot([pose_x[11], pose_x[23]], [pose_y[11], pose_y[23]], [pose_z[11], pose_z[23]], 'r-', alpha=0.6)
        ax.plot([pose_x[12], pose_x[24]], [pose_y[12], pose_y[24]], [pose_z[12], pose_z[24]], 'r-', alpha=0.6)



    
    # Plot left hand landmarks
    if unified_landmarks['left_hand']:
        left_x = [p['x'] for p in unified_landmarks['left_hand']]
        left_y = [p['y'] for p in unified_landmarks['left_hand']]
        left_z = [p['z'] for p in unified_landmarks['left_hand']]
        ax.scatter(left_x, left_y, zs=left_z, c='blue', marker='^', s=30, label='Left Hand')
    
    # Plot right hand landmarks
    if unified_landmarks['right_hand']:
        right_x = [p['x'] for p in unified_landmarks['right_hand']]
        right_y = [p['y'] for p in unified_landmarks['right_hand']]
        right_z = [p['z'] for p in unified_landmarks['right_hand']]
        ax.scatter(right_x, right_y, zs=right_z, c='green', marker='s', s=30, label='Right Hand')
    
    # Set labels and title
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Landmarks Visualization (Rotate with mouse)')
    ax.legend()
    
    # Equal aspect ratio
    ax.set_box_aspect([1, 1, 1])
    plt.show()

# Usage:
plot_3d_landmarks(main())
```
